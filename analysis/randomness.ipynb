```r
cal_ari_matrix <- function(data){
    # Create an empty similarity matrix
    num_elements <- dim(data)[1]
    ari_matrix <- matrix(0, nrow = num_elements, ncol = num_elements)
    ari2_matrix <- matrix(0, nrow = num_elements, ncol = num_elements)

    # Calculate pairwise similarities and fill the matrix
    for (i in 1:num_elements) {
    for (j in 1:num_elements) {
        a <- data$clustering_file[i]
        b <- data$clustering_file[j]
        ca <- read.table(a, header = T, sep="\t", comment.char = "")
        cb <- read.table(b, header = T, sep="\t", comment.char = "")
        ari_matrix[i, j] <- aricode::ARI(ca$clusterings, cb$clusterings)
        ari2_matrix[i, j] <- adjusted_wallance_indices(true=ca$clusterings, pred=cb$clusterings)$ARI2
    }
    }

    rownames(ari_matrix) <- data$seed
    colnames(ari_matrix) <- data$seed
    rownames(ari2_matrix) <- data$seed
    colnames(ari2_matrix) <- data$seed
    return(list(ari=ari_matrix, ari2=ari2_matrix))
}
```

```r
hm_ari <- function(data, data_matrix, hm_color, minv, a, b, name="ARI"){ # data: df for each method and each resolution
    # k_ls <- unique(data$n_clusters)
    # if(length(k_ls) == 1){
    #     a <- data$n_clusters[1]-1
    #     b <- data$n_clusters[1]+1
    # } else{
    #     a <- min(data$n_clusters)
    #     b <- max(data$n_clusters)
    # }
    ha_column1 = HeatmapAnnotation(k = anno_points(data$n_clusters, ylim=c(a,b), k_opt=anno_lines(7)), 
        annotation_name_side = "left",height=unit(3, 'cm'),show_annotation_name = FALSE)
    interval_width <- (1 - minv) / (length(hm_color)-1)
    seqc <- seq(minv, 1, by = interval_width)
    col_fun = colorRamp2(seqc, hm_color) 

    hm <- Heatmap(data_matrix, 
    name = name, 
    col = col_fun, 
    show_column_dend = FALSE,
    show_row_dend = FALSE,
    rect_gp = gpar(col = "black", lwd = 1), 
    top_annotation = ha_column1,
    show_heatmap_legend = FALSE,
    cell_fun = function(j, i, x, y, width, height, fill) {
        grid.text(sprintf("%.3f", data_matrix[i, j]), x, y, gp = gpar(fontsize = 10))},
    # bottom_annotation = column_ha, show_column_names = FALSE,
    # heatmap_legend_param = list(title = "MADs", at=c(-maxv, 0, maxv), labels=c("worst","median","best"), direction="horizontal" )
    )
    return(hm)
}
```

```r
w_ls <- c(12, 20, 12, 30, 20, 15)
len_ls <- c(5.05, 9.15, 5.05, 12.12, 8.12, 6.07)

i <- 2
#### load meta data
dataset <- dataset_ls[i]
df <- read.table(file="meta_info_candidate1.tsv", sep="\t", header=TRUE)
df_metrics <- read.table(file="metrics_value_candidate1.tsv", sep="\t", header=TRUE)
df$dataset <- dataset_name[i]
df_metrics$dataset <- dataset_name[i]
k_opt <- k_ls[i]

#### set work dir
setwd(paste0("/home/siluo/public/SiyuanLuo/projects/benchmark/outputs/", dataset))

#### calculate ari
minv <- 0.8
interval_width <- (1 - minv) / (length(hm_color)-1)
seqc <- seq(minv, 1, by = interval_width)
col_fun = colorRamp2(seqc, hm_color) 

width <- w_ls[i]
for(lm in unique(df$long_method)){
    ############### calculating ari ###################
    df_tmp <- df %>% filter(long_method == lm & seed %in% c(0, 2, 5, 42, 123))
    all_ari <- matrix(0, nrow = 5, ncol = 0)
    a <- min(unique(df_tmp$n_clusters[df_tmp$n_clusters!=0]))
    b <- max(unique(df_tmp$n_clusters))
    for(r in unique(df_tmp$resolution)){
        if(r !="latent_space"){
            df_sub <- df_tmp %>% filter(resolution == r)
            data <- df_sub[, c("clustering_file", "resolution", "seed", "n_clusters")]
            res <- cal_ari_matrix(data)
            all_ari <- cbind(all_ari, res$ari)
        }
    }
    ############### plotting ###################
    k_min <- min(c(df_tmp$n_clusters[df_tmp$n_clusters!=0], k_opt))
    k_max <- max(c(df_tmp$n_clusters[df_tmp$n_clusters!=0], k_opt))

    ha_column1 = HeatmapAnnotation(k = anno_points(df_tmp$n_clusters[df_tmp$n_clusters!=0], ylim=c(k_min, k_max)), 
        annotation_name_side = "left",height=unit(2, 'cm'))
    ht_opt$TITLE_PADDING = unit(c(8.5, 8.5), "points")
    hm <- Heatmap(all_ari, 
        name = lm, 
        col = col_fun, 
        show_column_dend = FALSE,
        show_row_dend = FALSE,
        rect_gp = gpar(col = "black", lwd = 1), 
        top_annotation = ha_column1,
        cluster_rows = FALSE, 
        cluster_columns = FALSE, 
        # show_heatmap_legend = FALSE,
        cell_fun = function(j, i, x, y, width, height, fill) {
            grid.text(sprintf("%.2f", all_ari[i, j]), x, y, gp = gpar(fontsize = 10))},
        # bottom_annotation = column_ha, 
        # show_column_names = FALSE,
        heatmap_legend_param = list(title = "ARI", at=c(minv, 1) ), 
        column_split = df_tmp$resolution[df_tmp$resolution!="latent_space"],
        row_title = lm,
        row_title_gp = gpar(fill = "grey", col = "black", border = "black", fontsize = 9)
        )


    pdf(paste0("/home/siluo/public/SiyuanLuo/projects/benchmark/figures/",lm, "_", dataset, ".pdf"), width=width, height=3.5)
    draw(hm)
    decorate_annotation("k", {
    grid.lines(c(0, len_ls[i]), unit(c(k_opt, k_opt), "native"), gp = gpar(col = "red",lty = 2))
    })
    dev.off() 
}

```
