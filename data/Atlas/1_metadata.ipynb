```python
%matplotlib inline
plt.figure(figsize = (50, 15))
sns.set(font_scale=1.8)
ax = sns.heatmap(df_percentage, 
                 annot=False, 
                 linewidths=1, 
                 linecolor='white',
                 cbar_kws={"pad": 0.02})


for ind, tissue in enumerate(tissue1):
    id_row = list(df_percentage.index).index(tissue)
    id_col = list(df_percentage.columns).index(type1[ind])
    ax.add_patch(plt.Circle((id_col+0.5, id_row+0.5), 0.4,  fc='none', ec='blue', lw=4, clip_on=False))

for ind, tissue in enumerate(tissue2):
    id_row = list(df_percentage.index).index(tissue)
    id_col = list(df_percentage.columns).index(type2[ind])
    ax.add_patch(plt.Rectangle((id_col, id_row), 1, 1, fc='none', ec='red', lw=4, clip_on=False))

ax.figure.tight_layout()

plt.savefig('clusters_tissue_composition.pdf')
# plt.savefig('clusters_tissue_composition.svg')
```

When $p_i=0$, Shannon entropy cannot count the an element. Add 0.01 to all $p$.

```python
df_percentage_e = df_percentage + 0.01
```

```python
# upper bound of entropy
math.log(111)
```

```python
entropy = df_percentage_e.apply(lambda x : scipy.stats.entropy(x.values), axis=0)
```
