```python
inputFiles = []
idxFiles = []
names = []

for i in range(pairs.shape[0]):
    tissue = pairs.loc[i, "tissue"]
    cluster = pairs.loc[i, "cell_type"]
    inputFiles.append(input_dir  + tissue + "_fragments_sorted_extracted_" + cluster + ".bed")
    idxFiles.append(input_dir  + tissue + "_fragments_sorted_extracted_" + cluster + ".cells.id.txt")
    names.append(tissue + "+" + cluster)
inputFiles   
```

```python
df_meta= pd.read_csv("~/projects/simulation/raw_data/bed/selected_cell/new_metadata.csv", sep=',', names=["cellID", "sample", "replicate", "logUMI", "tsse", "tissue", "cell type", "Life stage"])
df_meta["cell type"] = df_meta["cell type"].str.replace(" ", "_")
df_meta
```

```python
mask1 = df_meta.tissue.apply(lambda x: any(item for item in pairs.tissue.values if item in x))
mask2 = df_meta["cell type"].apply(lambda x: any(item for item in pairs.cell_type.values if item in x))
df_meta = df_meta[mask1]
df_meta = df_meta[mask2]
```
